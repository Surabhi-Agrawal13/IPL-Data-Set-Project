```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
IPL_data = pd.read_csv("IPL_Dataset.csv")
IPL_data.head()
```

### Q-1. What is the maximum number of matches played by an individual player in a season?

```python
Max_matches = IPL_data.groupby(['Player'])['Mat'].max().sort_values(ascending=False)
print("Maximum number of matches played by an individual player in the season : ", Max_matches)
```

### Q-2. What is the highest average of a player in a season?

```python
Highest_Avg = IPL_data["Avg"].max()
print("The highest average of a player in a season : ", Highest_Avg)
```

### Q3. What is the average number of matches played by the players in a season?

```python
Avg_matches = IPL_data["Mat"].mean()
print("The average number of matches played by the players in a season : ", Avg_matches)
```

### Q-4. Please check the correlation between all the features.

```python
columns_to_exclude = ['Player', 'HS']
data = IPL_data.drop(columns=columns_to_exclude)
correlation_matrix = data.corr()
print("Correlation Matrix:")
print(correlation_matrix)
#Heat Map for correlation
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

### Q-5. Check the list of players who has an average greater than 50 as well strike rate above 120.

```python
Player_list = IPL_data[(IPL_data['Avg'] > 50) & (IPL_data['SR'] > 120)]
List =  Player_list[['Player', 'Avg', 'SR']].reset_index()
print('The list of players who has an average greater than 50 as well strike rate above 120 : ')
print(List)
```

### Q-6. Please check the list of players who has an average greater than 40 and balls faced above 100.

```python
List_of_player = IPL_data[(IPL_data['Avg'] > 40) & (IPL_data['BF'] > 100)]
Avg_Ballfaced = List_of_player[['Player', 'Avg', 'BF']].reset_index()
print('The list of players who has an average greater than 40 and balls faced above 100: ')
print(Avg_Ballfaced)
```

### Q-7. Please check the list of players who scored more than 300 runs with an average greater than 45 & strike rate above 130.

```python
List_player = IPL_data[(IPL_data['Runs'] > 300) & (IPL_data['Avg'] > 45) & (IPL_data['SR'] > 130)]
Runs_Avg_SR = List_of_player[['Player', 'Runs', 'Avg', 'SR']].reset_index()
print('Below is the list of Players : ')
print(Runs_Avg_SR)
```

### Q-8. Players who scroed alteast one century in this season.

```python
Scored_century = IPL_data[IPL_data['100'] >=1]
sorted_players = Scored_century.sort_values(by='100', ascending=False).reset_index()
print('Players who scored at least one century in this season : ')
print(sorted_players[['Player', '100']])
```

### Q-9. Players who scored atleast 4 half centuries in this season.

```python
Scored_halfcentury = IPL_data[IPL_data['50'] >=4]
sorted_player = Scored_halfcentury.sort_values(by='50').reset_index()
print('Players who scored atleast 4 half centuries in this season : ')
print(sorted_player[['Player', '50']])
```

### Q-10. Check the list of players who hit more than 45 boundARIES and more than 10 sixes in this season.

```python
Highest_Four_Six = IPL_data[(IPL_data['4s'] > 45) & (IPL_data['6s'] > 10)]
Fours_Sixes = Highest_Four_Six[['Player', '4s', '6s']].reset_index()
print('The list of players who hit more than 45 boundaries and more than 10 sixes in this season : ')
print(Fours_Sixes)
```

### Q-11. Plot a histogram of number of matches played in a season by players.

```python
plt.hist(IPL_data['Mat'], bins=8, edgecolor='red', color='pink')
plt.title('Number of matches played')
plt.xlabel('Number of Matches')
plt.ylabel('Players')
plt.show()
```

### Q-12. Plot the histogram of balls faced by players.

```python
Ballfaced = plt.hist(IPL_data['Mat'], bins=8, edgecolor='yellow', color='blue')
plt.title('Ball Faced by Players')
plt.show()
```

### Q-13. Top 20 players with most runs in a season.

```python
top_20_players = IPL_data.sort_values(by='Runs', ascending=False).head(20)
print('Top 20 players with the most runs in the season:')
print(top_20_players[['Player', 'Runs']])
```

### Q-14. Bar chart for top 20 players with best strike rate in a season.

```python
top_20_players_sr = IPL_data.sort_values(by="SR", ascending=False).head(20)
plt.figure(figsize=(10, 6))
plt.bar(top_20_players_sr['Player'], top_20_players_sr['SR'], color='skyblue')
plt.xlabel('Player')
plt.ylabel('Strike Rate')
plt.title('Top 20 Players with the Best Strike Rate in a Season')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Q-15. Top 30 players with highest number of sixes.

### Q-16. Scatter plot of runs scored by a player v/s balls faced in a season.

```python
plt.scatter(IPL_data['Runs'], IPL_data['BF'], alpha=0.8, edgecolor='yellow', color='blue')
plt.title('Runs scored by a player v/s balls faced in a season')
plt.xlabel('Scored Runs')
plt.ylabel('Balls Faced')
plt.show()
```
